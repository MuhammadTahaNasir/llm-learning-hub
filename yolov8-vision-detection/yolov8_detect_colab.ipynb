```python
# ✅ Step 2 (Re-run after restart): Import the goodies
try:
    from ultralytics import YOLO
    from ultralytics.nn.tasks import DetectionModel
    from ultralytics.nn.modules.conv import Conv
    from torch.nn import Sequential
    import torch
    import cv2
    import matplotlib.pyplot as plt
    import numpy as np

    print("🛠️ Libraries imported — let’s get to work!")
except Exception as e:
    print(f"😕 Import failed: {e}")
    raise
```

```python
# ✅ Step 3: Load the YOLOv8 model

from ultralytics import YOLO

# Load the YOLOv8n model (now works since torch 2.5.1 is used)
model = YOLO("yolov8n.pt")
print("🚀 YOLOv8 model loaded — time to hunt for objects!")
```

```python
# ✅ Step 4: Run detection on a sample image
# Testing on a cool bus image from Ultralytics. Let’s see what YOLO finds!
try:
    image_path = "https://ultralytics.com/images/bus.jpg"
    results = model(image_path)
    print("🖼️ Detection done — objects found in the image!")
except Exception as e:
    print(f"😕 Image detection failed: {e}")
    raise
```

```python
# ✅ Step 5: Visualize the results
# Colab doesn’t like OpenCV’s show(), so we’ll use Matplotlib to display the detected image.
try:
    # Get the first result’s image with bounding boxes
    img = results[0].plot()  # YOLO’s built-in plot method adds boxes and labels
    plt.figure(figsize=(10, 8))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct colors
    plt.axis('off')  # Hide axes for a clean look
    plt.show()
    print("🎨 Detection image displayed — check out those boxes!")
except Exception as e:
    print(f"😕 Visualization failed: {e}")
    raise
```

```python
# ✅ Step 6: Save the detected image
# Saving the image with bounding boxes to your Colab files.
try:
    results[0].save(filename="detected.jpg")
    print("💾 Detected image saved as 'detected.jpg' — go check it out!")
except Exception as e:
    print(f"😕 Saving image failed: {e}")
    raise
```

```python
# ✅ Step 7: Export detection details
# Printing the detected objects’ classes and bounding boxes.
try:
    boxes = results[0].boxes
    print("\n🕵️‍♂️ Detections:")
    for cls, box in zip(boxes.cls, boxes.xyxy):
        print(f"Class {int(cls.item())} (Label: {model.names[int(cls.item())]}), Box: {box.tolist()}")
except Exception as e:
    print(f"😕 Exporting detections failed: {e}")
    raise
```
